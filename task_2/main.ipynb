**Ответы на вопросы:**
1. **Что за белок?** Поиск с помощью алгоритма BLASTP по базе данных `nr` показал, что данный фрагмент принадлежит белку **Родопсин (Rhodopsin)** (в частности, 100% совпадение с бычьим родопсином *Bos taurus*).
2. **Есть ли ортолог у человека и выбор инструмента:** Да, ортолог у человека есть — Human Rhodopsin (ген RHO).
   * **Выбор программы:** Лучше всего использовать **blastp**.
   * **Выбор базы данных:** `refseq_protein` (или `nr`), с ограничением по организму *Homo sapiens* (taxid: 9606).
   * **Обоснование:** Поскольку мы располагаем белковой последовательностью, поиск гомологов и ортологов на аминокислотном уровне (`blastp`) значительно чувствительнее и точнее, чем поиск по нуклеотидам (`tblastn` по транскриптому/геному) из-за вырожденности генетического кода. Геном человека отлично аннотирован, поэтому все его белки уже находятся в базах `refseq_protein` и искать по нуклеотидным базам излишне.

Мутации в этом белке связаны с развитием **пигментного ретинита** (Autosomal dominant retinitis pigmentosa, adRP) и **врожденной стационарной ночной слепоты**. Это приводит к потере фоторецепторов-палочек, что клинически выражается в потере сумеречного зрения (куриная слепота) и прогрессирующем сужении полей зрения (туннельное зрение).

### 3.2 Поиск кандидатов на белковую последовательность люциферазы

#### 1. Загрузка TransDecoder
Так как мы работаем локально, скачаем TransDecoder в текущую директорию с GitHub.

```python
%%bash

curl -L -o TransDecoder.tar.gz https://github.com/TransDecoder/TransDecoder/archive/refs/tags/TransDecoder-v5.7.1.tar.gz

tar -zxvf TransDecoder.tar.gz > /dev/null

mv TransDecoder-TransDecoder-v5.7.1 TransDecoder
echo "TransDecoder успешно установлен!"
```

#### 2. Запуск TransDecoder (Поиск ORF)
Используем `TransDecoder.LongOrfs` для поиска открытых рамок считывания (ORF) и перевода транскриптома в белковые последовательности.

```python
%%bash
./TransDecoder/TransDecoder.LongOrfs -t transcriptome.fasta
```

```python
%%bash
./TransDecoder/TransDecoder.Predict -t transcriptome.fasta
```

#### 3. Создание базы данных BLAST

```python
%%bash

makeblastdb -in transcriptome.fasta.transdecoder.pep -dbtype prot -out transcriptome_pep_db
```

#### 4. Поиск люциферазы с помощью BLAST и отбор топ 10 хитов

```python
%%bash

blastp -query luciferase_lc.fasta -db transcriptome_pep_db -outfmt 6 -max_target_seqs 10 > blast_results.txt

cat blast_results.txt
```

#### 5. Сохраняем лучшие последовательности

```python
%%bash

cut -f2 blast_results.txt > ids_list.txt

seqtk subseq transcriptome.fasta.transdecoder.pep ids_list.txt > best_luciferase_hits.fasta

echo "Сохранено в best_luciferase_hits.fasta. Вот их заголовки:"
grep ">" best_luciferase_hits.fasta
```
